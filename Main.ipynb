## Pandas time functions

```python
import pandas as pd
pd.date_range(start='2021-03-24', end='2021-09-01')
```

```python
pd.Series(pd.date_range("2021", freq="D", periods=3))
```

```python
df = pd.DataFrame({'year': [2021, 2022],
    'month': [3, 4],
    'day': [24, 25]}
)
ts1 = pd.to_datetime(df);print(ts1)
ts2 = pd.to_datetime('20210324', format='%Y%m%d'); print(ts2)
```

```python
import numpy as np
rng = pd.date_range('2021-03-24', '2021-09-01', freq='D')
ts = pd.Series(np.random.randn(len(rng)), index=rng); ts.head()
```
